```python
# num_cars.drop(columns = "normalized-losses", inplace = True)
num_cars.drop(labels = "normalized-losses", axis = 1, inplace = True)
num_cars.head()
```

```python
"""
nan_rows_cols = ['bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
nan_rows = (num_cars[nan_rows_cols].isnull().sum(axis = 0) > 0)
nan_rows
"""
clean_num_cars = num_cars.dropna(axis = 0)
clean_num_cars.info()
```

```python
obj_cols = ['bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
clean_num_cars[obj_cols].astype(dtype = "float", inplace = True)
clean_num_cars.info()
```

```python
norm_clean_num_cars = (clean_num_cars - clean_num_cars.min()) / (clean_num_cars.max() - clean_num_cars.min())
# Mieux vaut ne pas normaliser la colonne cible "price"
norm_clean_num_cars["price"] = clean_num_cars["price"]
norm_clean_num_cars.head(10)
```

```python
print(norm_clean_num_cars.columns)
print(type(norm_clean_num_cars.columns))
for col in norm_clean_num_cars.columns:
    print(type(col))
num_cols = ['symboling', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
print(num_cols == norm_clean_num_cars.columns)
```

```python
# DATAQUEST SOLUTION
# Select only the columns with continuous values from - https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names
# N.B. They keep the "normalized-losses" column but not the "symboling" one.
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars = numeric_cars.astype('float')
# Because `price` is the column we want to predict, let's remove any rows with missing `price` values.
numeric_cars = numeric_cars.dropna(subset=['price'])
# Replace missing values in other columns using column means.
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
# Normalize all columnns to range from 0 to 1 except the target column.
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
numeric_cars
```

```python
print(norm_clean_num_cars.index)
shuffled_index = np.random.permutation(norm_clean_num_cars.index)
print(shuffled_index)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
# num_cols = ['symboling', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
# Inutile car cette liste peut être appelée avec norm_clean_num_cars.columns (qui est certes de type pandas.indexes.base.Index au lieu de list mais sur lequel on peut itérer de même dans une boucle for)
def knn_train_test(train_cols, target_col, df):
    # Attention, train_cols est de type list (indispensable pour .fit et .predict), donc y veiller lors de l'appel de la fonction !
    # Le réglage du seed n'est pas nécessaire, j'essaie juste de comprendre les disparités de résultats avec le corrigé...
    np.random.seed(1)
    # df.reindex(index = np.random.permutation(195)) est incorrect car du fait des lignes supprimées, l'index n'est pas [0: 195] mais [0: 204] avec des valeurs manquantes.
    df.reindex(index = np.random.permutation(df.index))    
    # Number of rows: 195 --> split: 98 + 97
    train_set = df.iloc[:98]
    test_set = df.iloc[98:]
    knn = KNeighborsRegressor()
    knn.fit(train_set[train_cols], train_set[target_col])
    rmse = (mean_squared_error(test_set[target_col], knn.predict(test_set[train_cols]))) ** (1/2)
    return rmse
for col in norm_clean_num_cars.columns:
    print(col, ": ", knn_train_test([col], "price", norm_clean_num_cars))
# Idem with the Dataquest DataFrame
print("")
for col in numeric_cars.columns:
    print(col, ": ", knn_train_test([col], "price", numeric_cars))
```

Not very surpringly (except maybe for the 5th one), the five **best training data attributes** are, in ascending order of RMSE:
- **city-mpg** :  3900
- **horsepower** :  4326
- **highway-mpg** :  4526
- **curb-weight** :  5006
- **width** :  6445

```python
def knn_train_test(train_cols, target_col, df, k):
    df.reindex(index = np.random.permutation(df.index))    
    train_set = df.iloc[:98]
    test_set = df.iloc[98:]
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(train_set[train_cols], train_set[target_col])
    rmse = (mean_squared_error(test_set[target_col], knn.predict(test_set[train_cols]))) ** (1/2)
    return rmse
import matplotlib as mpl
import matplotlib.pyplot as plt
% matplotlib inline
plt.rc("font", size = 20)
# OU mpl.rc("font", size = 20)
fig = plt.figure(figsize = (20, 70))
i = 1
ks = [k for k in range(1, 10, 2)]
for col in norm_clean_num_cars.columns:
    rmses = []
    for k in range(1, 10, 2):
        rmses.append(knn_train_test([col], "price", norm_clean_num_cars, k))
    ax = fig.add_subplot(7, 2, i)
    ax.set_ylim(0, 16000)
    ax.set_title(col)
    ax.set_xlabel("k")
    ax.set_ylabel("RMSE")
    ax.scatter(ks, rmses, s = 400, c = rmses)
    i += 1
plt.show()
# Zoom sur les 5 meilleurs attributs
best_cols = ["city-mpg", "horsepower", "highway-mpg", "curb-weight", "width"]
fig = plt.figure(figsize = (20, 100))
i = 1
for col in best_cols:
    rmses = []
    for k in range(1, 10, 2):
        rmses.append(knn_train_test([col], "price", norm_clean_num_cars, k))
    ax = fig.add_subplot(5, 1, i)
    ax.set_title(col)
    ax.set_xlabel("k")
    ax.set_ylabel("RMSE")
    ax.scatter(ks, rmses, s = 400, c = rmses)
    i += 1
plt.show()
```

Let's note that the best efficiency of the model is achieved using only **the 3 (or 4) best features, that is: city-mpg, horsepower and highway-mpg.**

```python
def knn_train_test(train_cols, target_col, df, k):
    df.reindex(index = np.random.permutation(df.index))    
    train_set = df.iloc[:98]
    test_set = df.iloc[98:]
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(train_set[train_cols], train_set[target_col])
    rmse = (mean_squared_error(test_set[target_col], knn.predict(test_set[train_cols]))) ** (1/2)
    return rmse
ks = [k for k in range(1, 26)]
fig = plt.figure(figsize = (20, 60))
for num_ft in range(2, 5):
    rmses = []
    for k in ks:
        rmses.append(knn_train_test(best_cols[:num_ft], "price", norm_clean_num_cars, k))
    ax = fig.add_subplot(3, 1, num_ft - 1)
    ax.scatter(ks, rmses, s = 400, c = rmses)
    ax.set_xlabel("k")
    ax.set_ylabel("RMSE")
    ax.set_title(best_cols[:num_ft])
plt.show()
# WITH DATAQUEST'S DATAFRAME
DQ_best_cols = ['horsepower', 'width', 'curb-weight' , 'city-mpg' , 'highway-mpg']
fig = plt.figure(figsize = (20, 60))
for num_ft in range(2, 6):
    rmses = []
    for k in ks:
        rmses.append(knn_train_test(DQ_best_cols[:num_ft], "price", numeric_cars, k))
    ax = fig.add_subplot(3, 1, num_ft - 1)
    ax.scatter(ks, rmses, s = 400, c = rmses)
    ax.set_xlabel("k")
    ax.set_ylabel("RMSE")
    ax.set_title(DQ_best_cols[:num_ft])
plt.show()
```

**The optimal values of the hyperparameter k** are:
